**Les Bibliotheques :**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

**Sauvgarder le dataset :**

```python
data = pd.read_csv('real_drug_dataset.csv')
print(data)
```

```python
print("Montre les 5 premières lignes du cette dataset##")
print(data.head())

print("affiche les information de fichier##")
print(data.info())

print("affiche les statistiques de fichier##")
print(data.describe())

print("affiche les valeurs manquantes##")
print(data.isnull().sum())

print("affiche les valeurs dupliqués##")
print(data.duplicated().sum())

print("affiche les valeurs dupliqués##")
print(data.duplicated())
print(data.columns)
```

 **Nettoyage de donnee :**

```python
# ============ 2. NETTOYAGE DES DONNÉES ============
# Vérifier les incohérences dans les données catégorielles
print("\n=== VALEURS UNIQUES PAR COLONNE ===")
categorical_cols = ['Gender', 'Condition', 'Drug_Name', 'Side_Effects']
for col in categorical_cols:
    print(f"\n{col}: {data[col].unique()[:20]}")  # Afficher les 20 premières valeurs uniques
    print(f"Nombre de valeurs uniques : {data[col].nunique()}")
# Normaliser les valeurs catégorielles (supprimer espaces, standardiser la casse)
for col in categorical_cols:
    data[col] = data[col].str.strip().str.title()
# Vérifier les plages de valeurs numériques
print("\n=== PLAGES DE VALEURS NUMÉRIQUES ===")
numeric_cols = ['Age', 'Dosage_mg', 'Treatment_Duration_days', 'Improvement_Score']
for col in numeric_cols:
    print(f"{col}: Min={data[col].min()}, Max={data[col].max()}")
# Filtrer les valeurs aberrantes évidentes
# Âge entre 0 et 120 ans
data = data[(data['Age'] >= 0) & (data['Age'] <= 120)]
# Durée de traitement positive
data = data[data['Treatment_Duration_days'] > 0]
# Score d'amélioration entre 0 et 10
data = data[(data['Improvement_Score'] >= 0) & (data['Improvement_Score'] <= 10)]
# Dosage positif
data = data[data['Dosage_mg'] > 0]
# Vérifier les incohérences entre Condition et Drug_Name
print("\n=== DISTRIBUTION DES MÉDICAMENTS PAR CONDITION ===")
condition_drug_counts = data.groupby(['Condition', 'Drug_Name']).size().reset_index(name='Count')
print(condition_drug_counts)

# Standardiser les effets secondaires (regrouper les variations)
print("\n=== EFFETS SECONDAIRES LES PLUS COURANTS ===")
top_side_effects = data['Side_Effects'].value_counts().head(10)
print(top_side_effects)

# ============ 4. ENCODAGE DES VARIABLES CATÉGORIELLES ============
# Préparer l'encodage one-hot pour les colonnes catégorielles
print("\n=== ENCODAGE CATÉGORIEL ===")
categorical_cols = ['Gender', 'Condition', 'Drug_Name', 'Side_Effects']

# Créer des variables indicatrices
data_encoded = pd.get_dummies(data, columns=categorical_cols, prefix=categorical_cols, drop_first=True)
print(data_encoded)
# ============ 5. NORMALISATION DES VALEURS NUMÉRIQUES ============
scaler = StandardScaler()
numeric_columns = ['Age', 'Dosage_mg', 'Treatment_Duration_days', 'Improvement_Score']

# Garder une copie des valeurs originales
data_numeric_original = data[numeric_columns].copy()

# Appliquer la normalisation
data_numeric_scaled = pd.DataFrame(
    scaler.fit_transform(data[numeric_columns]),
    columns=[f"{col}_scaled" for col in numeric_columns]
)

# Combiner les données
data_final = pd.concat([data_encoded, data_numeric_scaled], axis=1)
# ============ 7. SAUVEGARDE ============
# Sauvegarder le dataset nettoyé
data_final.to_csv('real_drug_dataset_cleaned.csv', index=False)
print("\n=== SAUVEGARDE ===")
print("Dataset nettoyé sauvegardé sous 'real_drug_dataset_cleaned.csv'")

# Sauvegarder également une version avec encodage simple pour visualisation
data_simple = data.copy()
data_simple.to_csv('real_drug_dataset_simple_cleaned.csv', index=False)
print("Version simplifiée sauvegardée sous 'real_drug_dataset_simple_cleaned.csv'")
```

**Regression :**

```python
# ============ PRÉPARATION DES DONNÉES ============
x = data['Improvement_Score'].values.reshape(-1, 1)  # Reshape pour scikit-learn (1 feature)
y = data['Dosage_mg'].values
print("=== INFORMATIONS SUR LES DONNÉES ===")
print(f"x (Improvement_Score) - Min: {x.min()}, Max: {x.max()}, Mean: {x.mean():.2f}")
print(f"y (Dosage_mg) - Min: {y.min()}, Max: {y.max()}, Mean: {y.mean():.2f}")
print(f"Dimensions de x: {x.shape}")
print(f"Dimensions de y: {y.shape}")

# ============ SPLIT DES DONNÉES ============
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("\n=== INFORMATIONS SUR LE SPLIT ===")
print(f"Taille x_train : {x_train.shape}")
print(f"Taille x_test : {x_test.shape}")
print(f"Taille y_train : {y_train.shape}")
print(f"Taille y_test : {y_test.shape}")

# ============ ENTRAÎNEMENT DU MODÈLE ============
regressor = LinearRegression()
regressor.fit(x_train, y_train)

print("\n=== PARAMÈTRES DU MODÈLE ===")
print(f"Coefficient (pente) : {regressor.coef_[0]:.4f}")
print(f"Intercept (ordonnée à l'origine) : {regressor.intercept_:.4f}")

# Équation de la droite de régression
print(f"\nÉquation de régression : Dosage_mg = {regressor.coef_[0]:.4f} * Improvement_Score + {regressor.intercept_:.4f}")

# ============ PRÉDICTION ET ÉVALUATION ============
y_pred = regressor.predict(x_test)

print("\n=== ÉVALUATION DU MODÈLE ===")
print("Quelques prédictions vs réalité :")
for i in range(5):
  print(f"  Réel: {y_test[i]:.1f}, Prédit: {y_pred[i]:.1f}");

# ============ VISUALISATION ============
plt.figure(figsize=(15, 10))

# Graphique 1: Données et droite de régression
plt.subplot(2, 2, 1)
plt.scatter(x, y, alpha=0.5, s=20, label='Données')
ordonne = np.linspace(x.min(), x.max(), 1000).reshape(-1, 1)
plt.plot(ordonne, regressor.predict(ordonne), color='r', linewidth=2, label='Droite de régression')
plt.xlabel('Improvement_Score', fontsize=12)
plt.ylabel('Dosage_mg', fontsize=12)
plt.title('Relation entre Improvement_Score et Dosage_mg', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)

# Ajouter l'équation sur le graphique
equation_text = f'y = {regressor.coef_[0]:.4f}x + {regressor.intercept_:.4f}'
plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Graphique 2: Prédictions vs Valeurs réelles
plt.subplot(2, 2, 2)
plt.scatter(y_test, y_pred, alpha=0.5, s=30)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Ligne parfaite')
plt.xlabel('Valeurs réelles (Dosage_mg)', fontsize=12)
plt.ylabel('Prédictions (Dosage_mg)', fontsize=12)
plt.title('Prédictions vs Valeurs réelles', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Métriques d'évaluation :**

```python
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE;',metrics.mean_squared_error(y_test,y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('R²:',metrics.r2_score(y_test,y_pred))
```
